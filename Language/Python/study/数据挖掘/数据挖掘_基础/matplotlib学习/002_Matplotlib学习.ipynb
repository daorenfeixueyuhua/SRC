```python
# 需求：显示某个城市11点到12点小时之间每分钟的温度变化折线图，温度范围在15~18之间
import random
# 1、准备数据
x = range(60)
y_guiyang = [random.uniform(15, 18) for i in x]
y_bijie = [random.uniform(30, 45) for i in x]

# 2、创建画布
plt.figure(figsize=(20, 8), dpi=80)

# 3、绘制图像
plt.plot(x, y_guiyang, color='r', label='guiyang')
plt.plot(x, y_bijie, color='b', label='beijing')

# 显示图列
plt.legend()
# 修改x、y刻度
# 准备x的刻度说明
x_label = ['{0} min'.format(i)for i in x]
y_label = ['{0}C'.format(i) for i in range(15, 20)]
# plt.xticks(x[::5], x_label[::5])
# plt.yticks(range(15,20, 1), y_label)


# 添加网格显示
plt.grid(linestyle='--', alpha=0.5)

# 添加描述信息
plt.xlabel('time')
plt.ylabel('temp')
plt.title('xxxxxxxxxxxx')
# 4、显示图像
plt.show()
```
